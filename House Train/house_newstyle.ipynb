```python
#Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

print("Setup done")
```

```python
#Panggil data
data = pd.read_csv("C:/Users/visco/Downloads/Housing Prize/train.csv")
tdata= pd.read_csv("C:/Users/visco/Downloads/Housing Prize/test.csv")

y = data['SalePrice'] #tujuan -> SalePrice
X = data.drop('SalePrice') #apa aja selain SalePrice
```

```python
#Handling Missing Values
num_col = X.select_dtypes(include=np.number).columns #kolom number
cat_col = X.select_dtypes(exclude=np.number).columns #kolom category

isi_num_kosong = SimpleImputer(strategy='median') #isi kolom num kosong pake median
isi_cat_kosong = SimpleImputer(strategy='most_frequent') #isi kolom cat kosong pake nilai yang sering muncul

X[num_col] = isi_num_kosong.fit_transform(X[num_col])
X[cat_col] = isi_cat_kosong.fit_transform(X[cat_col])

tdata[num_col] = isi_num_kosong.transform(tdata[num_col])
tdata[cat_col] = isi_cat_kosong.transform(tdata[cat_col])
```

```python
#One-Hot_Decoding (Pecah jadi banyak kolom)
X = pd.get_dummies(X)
tdata = pd.get_dummies(tdata)
```

```python
#Kalau dimensi data test dengan data train beda, isi dengan nilai 0 pada data test
tdata = tdata.reindex(columns=X.columns, fill_value=0)
```
