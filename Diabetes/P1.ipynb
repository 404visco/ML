```python
#SETUP
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
data= pd.read_csv("C:/Users/visco/Downloads/playground-series-s5e12/train.csv")
tdata= pd.read_csv("C:/Users/visco/Downloads/playground-series-s5e12/test.csv")
print(data.shape)
data.head()
```

```python
#Lihat Daftar Kolom
for col in data.columns:
    print(f'{col}: {data[col].dtypes}')
    
```

```python
#Tentukan variabel X dan y
y= data['diagnosed_diabetes']
X= data.drop(['id','diagnosed_diabetes'], axis=1)
```

```python
#Presentase data
cat_col= X.select_dtypes(include= 'object').columns
num_col= X.select_dtypes(exclude= 'object').columns
presentase_kategori= len(cat_col) / len(X.columns) *100

cat_col=['gender', 'ethnicity', 'education_level', 'income_level',
       'smoking_status', 'employment_status']
print(f'{presentase_kategori}%')
```

```python
train_X,val_X,train_y,val_y= train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
model= CatBoostClassifier(iterations=2000,learning_rate=0.03,depth=8,cat_features=cat_col, loss_function="Logloss", verbose=False, early_stopping_rounds=100)
model.fit(train_X,train_y)
y_prob= model.predict_proba(val_X)[:,1]
roc= roc_auc_score(val_y,y_prob)
print(f'iter: {2000}, learn rate: {0.03}, deep: {8} --> {roc}')  
```
